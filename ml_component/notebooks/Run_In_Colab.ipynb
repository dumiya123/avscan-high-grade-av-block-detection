```python
# 1. Setup Environment
from google.colab import drive
import torch
import os
import shutil
import zipfile

drive.mount('/content/drive')

if torch.cuda.is_available():
    print(f"\n✅ GPU Accelerated: {torch.cuda.get_device_name(0)}")
else:
    print("\n❌ ERROR: GPU not detected. Go to Runtime -> Change Runtime Type and select T4 GPU.")
```

```python
# 2. Intelligent Project Deployment
# This step unzips safely using Python-native methods (robust against folder spaces).
ZIP_PATH = '/content/drive/MyDrive/AtrionNet_Implementation.zip'
EXTRACT_PATH = '/content/AtrionNet_Project'

if os.path.exists(ZIP_PATH):
    # Clear old data and extract using Python-native zipfile
    if os.path.exists(EXTRACT_PATH): shutil.rmtree(EXTRACT_PATH)
    os.makedirs(EXTRACT_PATH)
    
    print("📦 Extracting project archive...")
    with zipfile.ZipFile(ZIP_PATH, 'r') as zip_ref:
        zip_ref.extractall(EXTRACT_PATH)
    
    # Dynamically find the ml_component directory beneath the extraction point
    ml_root = None
    for root, dirs, files in os.walk(EXTRACT_PATH):
        if 'ml_component' in dirs:
            ml_root = os.path.join(root, 'ml_component')
            break
    
    if ml_root:
        os.chdir(ml_root)
        print(f"📂 Successfully entered Research Core: {os.getcwd()}")
        !pip install -r requirements.txt
    else:
        print(f"❌ ERROR: 'ml_component' folder not found anywhere inside {ZIP_PATH}.")
else:
    print(f"❌ ERROR: Could not find '{ZIP_PATH}' in your Drive. Did you name it correctly?")
```
