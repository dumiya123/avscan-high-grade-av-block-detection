```python
import sys
sys.path.insert(0, '../')

import numpy as np
from src.inference.predictor import AVBlockPredictor
```

## Load Predictor

```python
predictor = AVBlockPredictor(checkpoint='../checkpoints/best_model.pth')
```

## Run Inference

```python
# Load sample ECG
import h5py

with h5py.File('../data/processed/ecg_data.h5', 'r') as hf:
    ecg_signal = hf['record_0']['signal'][:]

# Predict
results = predictor.predict(ecg_signal, generate_report=True)

# Print diagnosis
print(f"Diagnosis: {results['diagnosis']['av_block_type']}")
print(f"Confidence: {results['diagnosis']['confidence']:.1%}")
```
